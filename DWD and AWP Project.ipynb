```python
#apply LabelEncoder in full_name and league, position and Current Club columns, to change in number
from sklearn.preprocessing import LabelEncoder 
```

```python
#create object of LabelEncoder class
le=LabelEncoder()
df['full_name']=le.fit_transform(df['full_name'])
df['league']=le.fit_transform(df['league'])
df['position']=le.fit_transform(df['position'])
df['Current Club']=le.fit_transform(df['Current Club'])
```
